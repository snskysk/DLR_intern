#ドイツ航空宇宙センターにインターンをした際に行っていたデータの平均調査やグラフ化の例

```python
#実験データの平均調査とグラフ化
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format="retina"
#実験データを変数に代入
data01=np.loadtxt("./b_n1.dat")
data02=np.loadtxt("./b_n2.dat")
data03=np.loadtxt("./b_n3.dat")

#各データのx(pressure)は一定であるためdata01を参照
x=data01[:,0]/100
#行列計算をして平均を計算
ave=(data01+data02+data03)/3

#y軸は（1/（2の14乗））で表示したいため、配列からy軸だけを取り出して割る
yave=ave[:,1]/2**14

plt.figure(figsize=(8, 6.5))
plt.rcParams["font.size"] = 22
plt.rcParams["xtick.labelsize"] = 12
plt.rcParams["ytick.labelsize"] = 15
plt.rcParams["legend.fontsize"] = 12
plt.grid()


plt.xlim(0.93,1.07)
plt.ylim(0.52,0.56)

plt.xlabel("pressure(P/P100kPa)")
plt.ylabel("normalized intensity")


plt.plot(x,yave,"b",marker="o",markersize=4)
plt.show()
print(ave)
```

#CSVファイルの操作

```python
%matplotlib inline
%config InlineBackend.figure_format="retina"
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = np.genfromtxt("b_n1.csv",delimiter=",", skip_header=1, dtype='float')

x=data[:,0]/100
y=data[:,1]/2**14

plt.figure(figsize=(8, 6.5))
plt.rcParams["font.size"] = 22
plt.rcParams["xtick.labelsize"] = 12
plt.rcParams["ytick.labelsize"] = 15
plt.rcParams["legend.fontsize"] = 12
plt.grid()

plt.xlim(0.93,1.07)
plt.ylim(0.52,0.56)

plt.xlabel("pressure(P/P100kPa)")
plt.ylabel("normalized intensity")

plt.plot(x,y,"b",marker="o",markersize=4)
plt.show()
```
